```python
# Count number of unique items exported by each of the reporter countries, store results in a dictionary
num_items = []
countries = []

for country in items_by_country['Reporter Countries'].unique():
    itemcount = len(items_by_country.loc[items_by_country['Reporter Countries']==country,'Item'].unique())
    num_items.append(itemcount)
    countries.append(country)
itemscount_by_country = dict(zip(countries,num_items))
```

```python
itemscount_by_country
```

### Count trading partners per country

```python
partners_by_country = trade.groupby(['Reporter Countries','Partner Countries','Unit']).sum()
```

```python
partners_by_country.head()
```

```python
partners_by_country = partners_by_country.reset_index()
```

```python
partners_by_country.head()
```

```python
reporter_countries = []
partners = []

for country in partners_by_country['Reporter Countries'].unique():
    partner = len(partners_by_country.loc[partners_by_country['Reporter Countries']== country,'Partner Countries'].unique())
    reporter_countries.append(country)
    partners.append(partner)
partnerscount_by_country = dict(zip(reporter_countries,partners))
```

```python
partnerscount_by_country
```

```python
# Sort dictionary by values in descending order
# return country name and value which has the maximum partner numbers and item numbers 
max_items = sorted(itemscount_by_country.items(),key=operator.itemgetter(1),reverse=True)[0]
print('Country with maximum exporting items:', max_items)

max_partners = sorted(partnerscount_by_country.items(),key=operator.itemgetter(1),reverse=True)[0]
print('Country with maximum partner numbers:', max_partners)
```
